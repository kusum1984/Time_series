## **Time resampling**

```python
df_tesla.resample(rule='A').min()
```

```python
#Max shares in a year
df_tesla.resample(rule='A').max()
```

```python
#plotting with respect to Open column
#Rule A is year end frequency
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
#Quartly frequency
df_tesla.resample(rule='QS').max()['Open'].plot()
```

```python
#Quartly frequency
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
#Business End frequency
df_tesla.resample(rule='BA').max()['High'].plot()
```

```python
#Business Quarter freqyency
df_tesla.resample(rule='BQS').max().plot
```

```python
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max().plot(figsize=(15,6))
```

```python
#Finding rolling mean/mean average
df_tesla['High'].rolling(10).max().head(20)
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head(30)
```

```python
df_tesla[['Open','Open:30 days rolling']].plot()
```
